```python
import json
import os
```

```python
work_path = 'C:\\Users\\Jake\\Desktop\\Training_라벨링데이터'
destination = 'C:\\Users\\Jake\\Desktop\\labels'
```

```python
cls_dic= {
    '기타': 0,
    '신문지': 1,
    '상자류': 2,
    '음료수곽': 3,
    '종이봉투': 4,
    '일회용음료수잔': 5,
    '페트병': 6,
    '음료수병': 7,
    '맥주병': 8,
    '소주병': 9,
    '음료수캔': 10,
    '커피캔': 11,
    '봉투': 12,
    '일회용덮개': 13,
    '과자봉지': 12,
    '박카스병': 7,
    '기타술병': 7,
    '맥주캔': 10,
}
```

```python
tree = os.walk(work_path)

for files in tree:
    for f in files[2]:
        if '.json' in f or '.Json' in f:
            try:    
                with open(files[0]+'\\'+f, 'rt', encoding="utf-8") as old_label: 
                    old_label = json.load(old_label)
            except:
                with open(files[0]+'\\'+f, 'rt') as old_label: # 오류회vl
                    old_label = json.load(old_label)
            rng = len(old_label['Bounding'])
            for i in range(rng):
                try:
                    f_name = old_label['FILE NAME']
                    f_name = f_name.split('.')[0]
                    
                    csl_name = old_label['Bounding'][i]['DETAILS'] # 클래스 

                    x1 = old_label['Bounding'][i]['x1'] #x1,x2, y1,y2 값을 불러옴
                    x1 = int(x1)
                    y1 = old_label['Bounding'][i]['y1']
                    y1 = int(y1)
                    x2 = old_label['Bounding'][i]['x2']
                    x2 = int(x2)
                    y2 = old_label['Bounding'][i]['y2']
                    y2 = int(y2)

                    res = old_label['RESOLUTION'] # 이미지 사이즈를 가져옴 ( 이미지사이즈가 다다름 )
                    res = res.split('*')
                    x_res, y_res = res
                    x_res, y_res = int(x_res), int(y_res)

                    if x_res >= y_res:   # Padding이 될시에 x, y 길이에따라서 패딩그려지는곳이 달라짐.
                        halfpad = (x_res - y_res)/2
                        y1, y2 = y1 + halfpad, y2 + halfpad
                        y_res = x_res

                    elif x_res <= y_res:
                        halfpad = (y_res - x_res)/2
                        x1, x2 = x1 + halfpad, x2 + halfpad
                        x_res = y_res

                    x_center = ((x1+x2)/2)/x_res # x , y  bounding box의 중앙점 값
                    y_center = ((y1+y2)/2)/y_res

                    width = (x2-x1)/x_res # 패딩까지 고려해서 w ,h 값
                    height = (y2-y1)/y_res

                    cls_num = cls_dic[csl_name]

                    new_label_content = f'{cls_num} {x_center} {y_center} {width} {height}'

                    if i == 0:
                        with open(destination + '\\' + f_name + '.txt','at') as new_label:
                            new_label.write(new_label_content)
                    elif i > 0:
                        with open(destination + '\\' + f_name + '.txt','at') as new_label:
                            new_label.write('\n' + new_label_content)
                except Exception as e:
                    print(f)
                    print(e)
                    print()
```
